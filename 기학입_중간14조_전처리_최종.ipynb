```python
import matplotlib.pyplot as plt

def plot_avg_speeds_for_road(road_name, avg_speeds_df):
    road_data = avg_speeds_df[avg_speeds_df['도로명'] == road_name]

    plt.figure(figsize=(12, 6))
    for direction in road_data['방향'].unique():
        subset = road_data[road_data['방향'] == direction]
        plt.plot(subset.columns[2:], subset.iloc[0, 2:], label=f"direction: {direction}")

    plt.title(f"{road_name}_average_velocity")
    plt.xlabel("time")
    plt.ylabel("velocity")
    plt.legend()
    plt.show()
```

```python
road_names = merged_df['도로명'].unique().tolist()
road_names
```

```python
for road in road_names:
    plot_avg_speeds_for_road(road, avg_speeds)
```

```python
def plot_individual_speed_difference(avg_speeds_df):
    avg_speeds_df = avg_speeds_df.set_index(['도로명', '방향'])

    for road in road_names:
        if ('up' in avg_speeds_df.loc[road].index) and ('down' in avg_speeds_df.loc[road].index):
            diff = avg_speeds_df.loc[(road, 'up')].values - avg_speeds_df.loc[(road, 'down')].values

            # 막대그래프로 차이 시각화
            plt.figure(figsize=(10, 6))
            plt.bar(avg_speeds_df.columns, diff, color='skyblue')

            plt.xlabel('Time')
            plt.ylabel('Speed Difference (up - down)')
            plt.title(f'Average Speed Difference between up and down for {road}')
            plt.tight_layout()
            plt.show()

plot_individual_speed_difference(avg_speeds)
```

```python
def get_top3_roads_by_9am_difference(avg_speeds_df):
    avg_speeds_df = avg_speeds_df.set_index(['도로명', '방향'])
    speed_diff_9am = {}

    for road in road_names:
        if ('up' in avg_speeds_df.loc[road].index) and ('down' in avg_speeds_df.loc[road].index):
            # 절댓값 사용하여 속도 차이 계산
            diff_9am = abs(avg_speeds_df.loc[(road, 'up'), '9시평균속도'] - avg_speeds_df.loc[(road, 'down'), '9시평균속도'])
            speed_diff_9am[road] = diff_9am

    # 상위 3개 도로명 정렬
    top3_roads = sorted(speed_diff_9am, key=speed_diff_9am.get, reverse=True)[:3]
    return top3_roads

top3_roads = get_top3_roads_by_9am_difference(avg_speeds)
print("Top 3 roads with the highest absolute average speed difference at 9AM:", top3_roads)
```
